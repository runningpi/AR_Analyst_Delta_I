```python
import os

def list_files_in_folder(folder_path):
    """Gibt eine Liste aller Dateinamen im angegebenen Ordner zurck."""
    return [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

# Beispielaufruf:
ordner = "AMD/Advanced Micro Devices Inc_105_Technology & Electronics/"   # Pfad anpassen
dateien = list_files_in_folder(ordner)
print(dateien)
```
